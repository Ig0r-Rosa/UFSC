```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definição da função de transferência H(s) = 62831 / (s + 62831)
num = [62831]
den = [1, 62831]

# Criação do sistema de função de transferência
sistema = signal.TransferFunction(num, den)

# Frequências de análise (de 1 Hz a 100 MHz)
w, mag, phase = signal.bode(sistema, np.logspace(0, 8, 1000))

# Conversão de rad/s para Hz
frequencias_hz = w / (2 * np.pi)

# Cálculo da frequência de corte (ponto onde a magnitude é -3 dB)
magnitude_minima = min(mag)
indice_frequencia_corte = np.where(mag <= -3)[0][0]
frequencia_corte = frequencias_hz[indice_frequencia_corte]
print(f"Frequência de corte: {frequencia_corte:.2f} Hz")

# Cálculo do ganho em dB e linear
ganho_maximo_db = max(mag)
ganho_maximo_linear = 10 ** (ganho_maximo_db / 20)
print(f"Ganho máximo em dB: {ganho_maximo_db:.2f} dB")
print(f"Ganho máximo em linear: {ganho_maximo_linear:.2f}")

# Plot do diagrama de magnitude em Hz
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.semilogx(frequencias_hz, mag)
plt.title('Diagrama de Bode - Magnitude e Fase')
plt.ylabel('Magnitude (dB)')
plt.grid(which='both', linestyle='--', linewidth=0.5)

# Plot do diagrama de fase em Hz
plt.subplot(2, 1, 2)
plt.semilogx(frequencias_hz, phase)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Fase (graus)')
plt.grid(which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
